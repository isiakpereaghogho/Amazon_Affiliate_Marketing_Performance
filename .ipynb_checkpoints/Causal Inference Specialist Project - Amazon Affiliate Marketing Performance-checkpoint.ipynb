### INTRODUCTION & OVERVIEW

This comprehensive dataset provides real-world insights into Amazon Affiliate marketing performance, featuring detailed analytics on user behavior, product conversions, and revenue optimization strategies. Perfect for data scientists, marketing analysts, and e-commerce professionals looking to understand and improve affiliate marketing performance.

#### PROJECT GOALS

The Primary goal of the project is based on the question below: 

#### What is the causal effect of traffic source, product page exposure, seasonality, and user type (new vs returning) on conversion probability and affiliate revenue in an e-commerce affiliate platform?

This project also covers the following causal inference analysis questions:

Marketing Analytics
- Conversion Rate Optimization: How does exposure to high-performing product pages causally increase conversion rates relative to other pages?
- Customer Journey Mapping: What is the causal impact of early-stage user actions (first click source or landing page) on eventual purchase completion?
- Attribution Modeling: What is the causal contribution of each traffic source (organic, paid, social, referral) to conversions, compared to a counterfactual scenario where that source was absent?
- Seasonal Trend Analysis: What is the causal effect of seasonal periods (e.g., holiday vs non-holiday) on conversion rates across different product categories?

Business Intelligence
- Revenue Forecasting: How does an increase in user traffic from specific sources causally affect affiliate revenue over time?
- Product Performance: What is the causal effect of promoting specific product categories on revenue and conversion outcomes?
- User Segmentation: How does being a returning customer (versus a new customer) causally affect conversion probability and average order value?
- Geographic Analysis: What is the causal effect of user geographic location on conversion rates, after controlling for traffic source and product exposure?

Causal Churn Analysis
- Traffic Source & Churn: What is the causal effect of traffic source on the probability of user churn within 30 days?
- Early Engagement & Churn: How does early-session engagement causally affect the likelihood of user churn?
- Product Exposure & Churn: Does exposure to discounted or highly rated products causally reduce churn?

#### Conceptual DAG for the project based on the primary Causal Question

Key nodes

##### Treatments

- Traffic Source
- Product Category / Discount
- Seasonality
- User Type (New vs Returning)

##### Confounders

- Device Type
- Country / Geography
- Product Price
- Rating / Reviews
- User History (CLV, previous orders)

##### Outcomes

- Conversion
- Order Value
- Commission Earned

##### Post-treatment

- Funnel Stage
- Engagement Score
- Conversion Time


```python
from graphviz import Digraph

dag = Digraph(format="png")
dag.attr(rankdir="LR", size="8,10")

# ===== Treatments =====
dag.node("TrafficSource", "Traffic Source")
dag.node("ProductExposure", "Product Category / Discount")
dag.node("Seasonality", "Seasonality")
dag.node("UserType", "New vs Returning User")

# ===== Confounders =====
dag.node("Device", "Device Type")
dag.node("Country", "Geography")
dag.node("Price", "Product Price")
dag.node("Rating", "Product Rating")
dag.node("Reviews", "Review Count")
dag.node("UserHistory", "User History (CLV, Prior Orders)")

# ===== Post-Treatment Variables =====
dag.node("Engagement", "Engagement Metrics\n(Time on Page, Scroll Depth)", style="dashed")
dag.node("FunnelStage", "Conversion Funnel Stage", style="dashed")

# ===== Outcomes =====
dag.node("Conversion", "Conversion")
dag.node("Revenue", "Affiliate Revenue")

# ===== Causal Edges =====
# Treatments → Outcomes
dag.edge("TrafficSource", "Conversion")
dag.edge("ProductExposure", "Conversion")
dag.edge("Seasonality", "Conversion")
dag.edge("UserType", "Conversion")

dag.edge("Conversion", "Revenue")

# Confounders → Treatments
dag.edge("Device", "TrafficSource")
dag.edge("Country", "TrafficSource")
dag.edge("UserHistory", "UserType")

dag.edge("Price", "ProductExposure")
dag.edge("Rating", "ProductExposure")
dag.edge("Reviews", "ProductExposure")

# Confounders → Outcomes
dag.edge("Device", "Conversion")
dag.edge("Country", "Conversion")
dag.edge("Price", "Conversion")
dag.edge("Rating", "Conversion")
dag.edge("Reviews", "Conversion")
dag.edge("UserHistory", "Conversion")

# Post-treatment (do NOT control)
dag.edge("TrafficSource", "Engagement")
dag.edge("ProductExposure", "Engagement")
dag.edge("Engagement", "Conversion")

dag.edge("Conversion", "FunnelStage")

dag
```

I conducted dataset-level EDA to validate structure, timing, and causal roles before merging. Each dataset was audited for integrity, treatment timing, and post-treatment variables to prevent causal leakage.

```python
# Canonicalize Products Catalog
products_catalog_canonical = (
    products_catalog
    .groupby("product_asin", as_index=False)
    .agg({
        "product_title": "first",
        "brand": "first",
        "category": "first",
        "price": "median",
        "discount_percentage": "median",
        "rating": "mean",
        "review_count": "max",
        "commission_rate": "first",
        "bestseller_rank": "min",
        "seasonal_trend": "first"
    })
)

# Merge clicks with products
clicks_products = clicks.merge(
    products_catalog_canonical,
    on="product_asin",
    how="left"
    # validate="many_to_one"  # optional; skip for EDA
)


# Aggregate user_behaviour to session-level (pre-treatment)
# This ensures one row per session_id + user_id for safe merge
user_behaviour_agg = (
    user_behaviour
    .groupby(["session_id", "user_id"], as_index=False)
    .agg({
        "traffic_source": "first",
        "device_type": "first",
        "new_vs_returning": "first",
        "time_on_page_seconds": "mean",
        "scroll_depth_percentage": "mean",
        "user_engagement_score": "mean",
        "conversion_funnel_stage": "first",  # only for EDA
        "page_type": "first"  
    })
)

#Merge clicks+products with user_behavior
clicks_products_sessions = clicks_products.merge(
    user_behaviour_agg,
    on=["session_id", "user_id"],
    how="left"
)

#final merge: conversions
full_df = clicks_products_sessions.merge(
    conversions,
    on=["click_id", "user_id"],
    how="left"
)
```

```python
#quick sanity checks
print("Clicks:", len(clicks))
print("Clicks + products:", len(clicks_products))
print("Clicks + products + sessions:", len(clicks_products_sessions))
print("Full dataset:", len(full_df))
```
